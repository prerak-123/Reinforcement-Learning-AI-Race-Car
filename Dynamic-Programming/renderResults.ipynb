```python
import numpy as np
import matplotlib.pyplot as plt

policy = np.load("policy.npy")
value_function = np.load("value_function.npy")


fig = plt.figure(figsize=(6, 3.2))

ax = fig.add_subplot(111)
ax.set_title('Policy')
plt.imshow(policy)
ax.set_aspect('equal')

cax = fig.add_axes([0.12, 0.1, 0.78, 0.8])
cax.get_xaxis().set_visible(False)
cax.get_yaxis().set_visible(False)
cax.patch.set_alpha(0)
cax.set_frame_on(False)
plt.colorbar(orientation='vertical')
ax.set_xlabel("Cars at Second Location")
ax.set_ylabel("Cars at First Location")

plt.savefig("policy.png")
plt.show()
```

```python
fig2 = plt.figure(figsize=(6, 3.2))

ax2 = fig2.add_subplot(111)
ax2.set_title('Value Function')
plt.imshow(value_function)
ax2.set_aspect('equal')

cax2 = fig2.add_axes([0.12, 0.1, 0.78, 0.8])
cax2.get_xaxis().set_visible(False)
cax2.get_yaxis().set_visible(False)
cax2.patch.set_alpha(0)
cax2.set_frame_on(False)
plt.colorbar(orientation='vertical')
ax2.set_xlabel("Cars at Second Location")
ax2.set_ylabel("Cars at First Location")

plt.savefig("value_function.png")
plt.show()
```
